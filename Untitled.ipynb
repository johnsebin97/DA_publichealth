```python
# function to extract data
def extract_dataframe(data_frame, parameter, parameter_value):
    return data_frame.loc[data_frame[parameter] == parameter_value]
```

```python
# extract individuals having a High school degree and are between age 18-25
df_1825_Highschool = extract_dataframe(extract_dataframe(df, 
                                                        "Age", 
                                                        '18-25'), 
                                      'Educational Background', 
                                      "High School or Below")
```

```python
# extract individuals having a High school degree and are between age 26-40
df_2640_Highschool = extract_dataframe(extract_dataframe(df, 
                                                        "Age", 
                                                        '26-40'), 
                                      'Educational Background', 
                                      'High School or Below')
```
